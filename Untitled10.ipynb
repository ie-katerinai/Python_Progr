```python
import pandas as pd


df = pd.read_csv('sample_data/california_housing_train.csv')
import numpy as np
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data
```

```python
new_data = data.copy()
```

```python
new_data['robot'] = np.where(new_data['whoAmI']=='robot', 1, 0)
new_data['human'] = np.where(new_data['whoAmI']=='human', 1, 0)
```

```python
new_data.drop(columns=['whoAmI'],axis=1,inplace=True)
```
